```python
en2th_bpe2bpe = TransformerModel.from_pretrained(
                    model_name_or_path='mt/SCB_1M+TBASE_en-th_spm-spm_32000-joined_v1.0/models/',
                    checkpoint_file='checkpoint.pt',
                    data_name_or_path='mt/SCB_1M+TBASE_en-th_spm-spm_32000-joined_v1.0/vocab/',
                    bpe='sentencepiece',
                    sentencepiece_vocab='mt/SCB_1M+TBASE_en-th_spm-spm_32000-joined_v1.0/bpe/spm.en.model'
)
trans = en2th_bpe2bpe.translate
```

```python
import json
import copy
from google.colab import drive
drive.mount('/content/drive')
#!ls && cd "drive/MyDrive/Colab Notebooks/" && ls
squad = open('drive/MyDrive/Colab Notebooks/dev-v2.0.json')
squad_json = json.load(squad)
data = squad_json['data']
result = copy.deepcopy(data)
```

```python
print(data[0]['paragraphs'][0])
for i in range(len(data[0]['paragraphs'])):
    # data[0]['paragraphs'][i] is previously known as item
    for j in range(len(data[0]['paragraphs'][i]['qas'])):
        # data[0]['paragraphs'][i]['qas'][j] is previously known as question
        # print(str(data[0]['paragraphs'][i]['qas'][j]['question']))
        result[0]['paragraphs'][i]['qas'][j]['question'] = trans(str(data[0]['paragraphs'][i]['qas'][j]['question']))
        for k in range(len(data[0]['paragraphs'][i]['qas'][j]['answers'])):
            # data[0]['paragraphs'][i]['qas'][j]['answers'][k] is previously known as answer
            # print(str(data[0]['paragraphs'][i]['qas'][j]['answers'][k]['text']))
            result[0]['paragraphs'][i]['qas'][j]['answers'][k]['text'] = trans(str(data[0]['paragraphs'][i]['qas'][j]['answers'][k]['text']))
        try:
            for k in range(len(data[0]['paragraphs'][i]['qas'][j]['plausible_answers'])):
                # data[0]['paragraphs'][i]['qas'][j]['plausible_answers'][k] is previously known as answer
                # print(str(data[0]['paragraphs'][i]['qas'][j]['plausible_answers'][k]['text']))
                result[0]['paragraphs'][i]['qas'][j]['plausible_answers'][k]['text'] = trans(str(data[0]['paragraphs'][i]['qas'][j]['plausible_answers'][k]['text']))
        except:
            pass
    result[0]['paragraphs'][i]['context'] = trans(str(data[0]['paragraphs'][i]['context']))
    break
```

```python
translation = []
for item in result[0]['paragraphs']:
    for question in item['qas']:
        translation.append(question['question'])
        for answer in question['answers']:
            translation.append(answer['text'])
            translation.append(answer['answer_start'])
        try:
            for answer in question['plausible_answers']:
                translation.append(answer['text'])
                translation.append(answer['answer_start'])
        except:
            pass
    translation.append(item['context'])
    print("\n".join([str(s) for s in translation]))
    
    break
```
